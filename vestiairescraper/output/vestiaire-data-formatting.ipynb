```python
import json

gender_dict = {
    "Vestiaire-mens-accessories.json": "m",
    "Vestiaire-mens-bottoms.json": "m",
    "Vestiaire-mens-shoes.json": "m",
    "Vestiaire-mens-tops.json": "m",
    "Vestiaire-womens-bottoms.json": "f",
    "Vestiaire-womens-shoes.json": "f",
    "Vestiaire-mens-tops.json": "f"
}

category_dict = {
    "Vestiaire-mens-accessories.json": "Accessories",
    "Vestiaire-mens-bottoms.json": "Bottoms",
    "Vestiaire-mens-shoes.json": "Shoes",
    "Vestiaire-mens-tops.json": "Tops",
    "Vestiaire-womens-bottoms.json": "Bottoms",
    "Vestiaire-womens-shoes.json": "Shoes",
    "Vestiaire-mens-tops.json": "Tops"
}

source_jsons = ["Vestiaire-mens-accessories.json", "Vestiaire-mens-bottoms.json", "Vestiaire-mens-shoes.json",
                "Vestiaire-mens-tops.json", "Vestiaire-womens-bottoms.json", "Vestiaire-womens-shoes.json",
                "Vestiaire-mens-tops.json"]

with open("final-vestiaire-data.json", "w") as final_file:

    processed_jsons = []
    
    for source_json in source_jsons:

        with open(f"{source_json}", "r", encoding="utf-8") as json_file:
            loaded_jsons = json.load(json_file)

            gender = gender_dict[source_json]
            category = category_dict[source_json]
                
            for idx, article in enumerate(loaded_jsons):

                if article['title'] is None or article['price'] is None:
                    continue

                data = {
                    "ID": idx,
                    "name": article['title'],
                    "description": article['detailed']['description'],
                    "category": category,
                    "brand": article['detailed']['brand'],
                    "gender": gender,
                    "price": article['price'],
                    "prodLink": article['url'],
                    "prodImgLink": article['detailed']['image']['image']
                }

                processed_jsons.append(data)

    final_file.write(json.dumps(processed_jsons, indent=4))
```
