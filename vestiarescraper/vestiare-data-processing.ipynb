```python
import json

with open("vestiaire-data.json", "r", encoding="utf-8") as f:
    vestiare_jsons = json.load(f)

prod_types = []
[prod_types.append(j['product_type'].lower()) for j in vestiare_jsons if j['product_type'].lower() not in prod_types]
for prod_type in prod_types:
    print(prod_type)
```

```python
word_counts = {}

for prod_type in prod_types:
    tokens = prod_type.split()
    for token in tokens:
        if token in word_counts:
            word_counts[token] = word_counts[token] + 1
        else:
            word_counts[token] = 1

sorted_counts = dict(sorted(word_counts.items(), key=lambda x : x[1], reverse=True))
print(sorted_counts)
```

```python
top_keywords = ['dress', 'jacket', 'vest', 'blazer', 'coat', 'suit', 'top', 'polo', 
                'blouse', 'jumper', 'corset', 'cardigan', 'shirt', 'tunic', 't-shirt', 
                'jumpsuit', 'one-piece', 'swearshirt', 'jersey', 'peacoat', 'camisole', 
                'slip', 'parka', 'bra', 'knitwear', 'two-piece', 'one-piece', 'caban']

top_categories = []
for keyword in top_keywords:
    [top_categories.append(prod_type) for prod_type in prod_types if keyword in prod_type]

print("TOPS:")
for top in top_categories:
    print(top)

rest = [prod_type for prod_type in prod_types if prod_type not in top_categories]
print("\nREST:")
for prod_type in rest:
    print(prod_type)
```

```python
word_counts = {}

for prod_type in rest:
    tokens = prod_type.split()
    for token in tokens:
        if token in word_counts:
            word_counts[token] = word_counts[token] + 1
        else:
            word_counts[token] = 1

sorted_counts = dict(sorted(word_counts.items(), key=lambda x : x[1], reverse=True))
print(sorted_counts)
```

```python
bottom_keywords = ['skirt', 'pants', 'trousers', 'jeans', 'shorts', 'chino', 'leggings']

bottom_categories = []
for keyword in bottom_keywords:
    [bottom_categories.append(prod_type) for prod_type in rest if keyword in prod_type]

print("BOTTOMS:")
for bottom in bottom_categories:
    print(bottom)

rest = [prod_type for prod_type in rest if prod_type not in bottom_categories]
print("\nREST:")
for prod_type in rest:
    print(prod_type)
```

```python
accessories_categories = [prod_type for prod_type in prod_types if "lingerie" in prod_type]
print("ACCESSORIES:")
for access in accessories_categories:
    print(access)

rest = [prod_type for prod_type in rest if prod_type not in accessories_categories]
print("\nREST:")
for item in rest:
    print(item)
```

```python
top_categories.append('cashmere twin-set')
bottom_categories += ['bermuda', 'mini short', 'leather mini short']
accessories_categories.append('tight')

print("TOPS:")
for top in top_categories:
    print(top)

print("\nBOTTOMS:")
for bottom in bottom_categories:
    print(bottom)

print("\nACCESSORIES:")
for access in accessories_categories:
    print(access)
```
